```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simulation(S0 = 100,N = 100,steps = 10,K = 105):
    increments = np.random.uniform(-2,2,size = (N,steps))

    St = np.zeros((N,steps + 1))
    St[:,0] = S0

    for t in range(1,steps+1):
        St[:,t] = St[:,t-1] + increments[:,t-1]
    payoffs = np.maximum(St[:,-1] - K,0)
    return St, payoffs

St, payoffs = simulation()
print("Expected value of call option : ", payoffs.mean())

plt.figure(figsize=(10,6))
for i in range(100):
    plt.plot(St[i,:])

plt.show()
```
